```python
grid_rows = soup.find_all('div', class_='gridRow')

# Initialize a list to store all rows data
all_rows_data = []

# Iterate over each gridRow
for row in grid_rows:
    # Find all gridCell divs within the current row
    cells = row.find_all('div', class_='gridCell')
    # Extract data from each cell and strip whitespace
    row_data = [cell.text.strip() for cell in cells]
    # Append row data to the list of all rows data
    all_rows_data.append(row_data)

# Print all rows data
for row_data in all_rows_data:
    print(row_data)
        
```
